```python
import pandas as pd

def licz_sume_chunks(path, chunksize=50_000):
    total_sum = 0


    for chunk in pd.read_csv(path, chunksize=chunksize):
        total_sum += chunk["amount"].sum()

    return total_sum
```

```python
def licz_sume_generator(filename):
    total_sum = 0
    with open(filename, "r", encoding="utf-8") as f:
        header = next(f)
        for line in f:
            values = line.strip().split(",")
            amount = float(values[1])
            total_sum += amount
    return total_sum
```
