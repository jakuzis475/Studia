```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# --- 1. Stworzenie własnego małego zbioru danych (filmów) ---
data = {
    "Film": ["Avengers", "Joker", "Frozen II", "Star Wars", "Lion King", "Toy Story 4", "It Chapter Two", "Captain Marvel", "Aladdin", "Spider-Man"],
    "Domestic": [858373000, 107425131, 477373578, 296000000, 543638043, 434038008, 211631679, 426829839, 355559350, 390532085],
    "International": [1132150000, 53800000, 972430000, 677000000, 1400000000, 580000000, 321000000, 550000000, 504000000, 741000000]
}

df = pd.DataFrame(data)

# --- 2. Podstawowe informacje i statystyki ---
print("Informacje o zbiorze:")
print(df.info())
print("\nStatystyki opisowe:")
print(df.describe())

# --- 3. Analizy statystyczne ---
print("\nŚredni Domestic:", df['Domestic'].mean())
print("Najwyższy Domestic:", df['Domestic'].max())
print("Najniższy Domestic:", df['Domestic'].min())

correlation = df['Domestic'].corr(df['International'])
print("Korelacja Domestic-International:", correlation)

# --- 4. Wykres: Box Office według filmu ---
plt.figure(figsize=(10,6))
sns.barplot(x='Domestic', y='Film', data=df.sort_values('Domestic', ascending=False))
plt.title("Box Office (Domestic) – Top Filmów")
plt.xlabel("Box Office [USD]")
plt.ylabel("Film")
plt.show()

# --- 5. Raport w Markdown (do wklejenia w notebook) ---
"""
Interpretacja wykresu:
- Najwięcej zarobiły filmy superbohaterskie i familijne, np. Avengers i Frozen II.
```
