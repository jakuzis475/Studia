```python
import pandas as pd


data = {
    "Imię": ["Ala", "Bartek", "Celina"],
    "Wiek": [22, 25, 20],
    "Ocena": [4.5, 3.8, 5.0]
}

df = pd.DataFrame(data)
print("Oryginalny DataFrame:")
display(df)

print("\nStatystyki opisowe:")
print(df.describe())


df["Rok studiów"] = [2, 3, 1]
print("\nDataFrame po dodaniu kolumny 'Rok studiów':")
display(df)

srednia_ocena = df["Ocena"].mean()
print("\nŚrednia ocena:", srednia_ocena)


wysokie_oceny = df[df["Ocena"] > 4]
print("\nWiersze z oceną większą niż 4:")
display(wysokie_oceny)
```
